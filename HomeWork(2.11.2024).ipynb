<h1>Описание задачи и признаков датасета</h1> 
<h2>Описание задачи:</h2>
Нужно предсказать, отменит ли клиент бронирование, используя признак is_canceled в качестве целевой переменной.

<h2>Описание признаков из датасета hotel-booking-demand:</h2>


1. `hotel` — тип отеля: City Hotel или Resort Hotel.

2. `is_canceled` — отменено ли бронирование: 1 — отменено, 0 — нет.

3. `lead_time` — время между бронированием и заездом (в днях).

4. `arrival_date_year` — год заезда.

5. `arrival_date_month` — месяц заезда.

6. `arrival_date_week_number` — номер недели заезда в году.

7. `arrival_date_day_of_month` — день заезда в месяце.

8. `stays_in_weekend_nights` — количество ночей на выходные (суббота и воскресенье).

9. `stays_in_week_nights` — количество ночей в будние дни.

10. `adults` — количество взрослых гостей.

11. `children` — количество детей.

12. `babies` — количество младенцев.

13. `meal` — тип заказа питания:
    - `Undefined` — без питания,
    - `BB` — завтрак,
    - `HB` — полупансион,
    - `FB` — полный пансион.

14. `country` — страна проживания клиента (код страны).

15. `market_segment` — сегмент рынка:
    - `Direct` — прямой,
    - `Corporate` — корпоративный,
    - `Online TA` — онлайн-агентство,
    - `Offline TA/TO` — офлайн агентство/оператор,
    - `Groups `— группа,
    - и т.д.

16. `distribution_channel` — канал дистрибуции:
    - `Direct` — прямой,
    - `Corporate` — корпоративный,
    - `TA/TO `— агентство/оператор,
    - и т.д.

17. `is_repeated_guest` — является ли клиент повторным гостем (1 — да, 0 — нет).

18. `previous_cancellations` — количество предыдущих отмен.

19. `previous_bookings_not_canceled` — количество бронирований, которые не были отменены.

20. `reserved_room_type` — тип комнаты, зарезервированный клиентом.

21. `assigned_room_type` — тип комнаты, предоставленный клиенту (может отличаться от зарезервированного).

22. `booking_changes` — количество изменений бронирования, внесённых клиентом.

23. `deposit_type` — тип депозита:
    - `No Deposit` — без депозита,
    - `Non Refund` — без возврата,
    - `Refundable` — возвратный депозит.

24. `agent` — идентификатор агента, сделавшего бронирование.

25. `company` — идентификатор компании, сделавшей бронирование (не используется из-за большого количества пропущенных значений).

26. `days_in_waiting_list` — количество дней, проведённых в списке ожидания.

27. `customer_type` — тип клиента:
    - `Contract` — контрактный клиент,
    - `Group` — группа,
    - `Transient` — временный,
    - `Transient-Party` — временная группа.

28. `adr` — средняя дневная ставка (average daily rate), стоимость номера в день.

29. `required_car_parking_spaces` — количество необходимых парковочных мест.

30. `total_of_special_requests` — количество специальных запросов от клиента.

31. `reservation_status` — статус бронирования:
    - `Canceled` — отменено,
    - `Check-Out` — завершено,
    - `No-Show` — не явился.

32. `reservation_status_date` — дата последнего изменения статуса бронирования.

<h2>Импортируем библиотеки:<h2>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h2>Считываем и выводим данные:</h2>

```python
data = pd.read_csv('hotel_bookings.csv',sep=',', engine='python')
```

<h2>Обрабатываем пропущенные значения:</h2>


В нашем датасете есть признак `company` который содержит 94% пропущенных значений, удалим его из таблицы:

```python
data = data.drop(columns=['company'])
data.head()
```

В датасете также есть признак `agent` который содержит 14% пропущенных значений, удалим все строки с пропущенными значениями из датасета.

```python
data = data.dropna(axis = 0)
data.head()
```

```python
print(data)
```

Теперь у нас в таблице 102894 строки и 31 столбец.

<h2>Обрабатываем признаки:</h2>

Разделим признаки на категориальные и количественные:

```python
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data.columns if data[c].dtype.name != 'object']
print(categorical_columns)
print(numerical_columns)
```

```python
for c in categorical_columns:
    print(c, data[c].unique())
```

Выведем информацию о наших категориальных признаках и занесем ее в отдельную переменную:

```python
data_describe = data.describe(include = [object])
data_describe
```

Разделим наши категориальные признаки на бинарные и небинарные:

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print(binary_columns, nonbinary_columns)
```

Обработаем единственный бинарный признак `hotel`:

```python
data['hotel'].unique()
```

```python
data.loc[data['hotel'] == 'Resort Hotel', 'hotel'] = 0
data.loc[data['hotel'] == 'City Hotel', 'hotel'] = 1
data['hotel'] = data['hotel'].astype(int)
data['hotel'].describe()
```

К небинарными признакам применим метод векторизации:

```python
data[nonbinary_columns]
```

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

```python
len(data_nonbinary.columns)
```

Проведем нормализацию количественных признаков (целевую переменную не нормализуем):

```python
data_numerical = data[numerical_columns]
data_numerical.describe()
```

```python
target = data_numerical['is_canceled']
data_features = data_numerical.drop(columns=['is_canceled'])
data_features = (data_features - data_features.mean(axis=0)) / data_features.std(axis=0)
data_numerical = data_features.copy()
data_numerical['is_canceled'] = target
data_numerical.describe()
```

Соединяем все в одну таблицу:

```python
data = pd.concat((data_numerical, data_nonbinary, data[binary_columns]), axis = 1)
print(data.shape)
```

```python
data.describe()
```

<h2>Задаем X и y:</h2>

```python
type(data)
```

```python
X = data.drop('is_canceled', axis = 1)
y = data['is_canceled']
feature_names = X.columns
```

```python
feature_names
```

```python
X = X.to_numpy()
y = np.array(data['is_canceled'])
y
```

```python
print(X.shape)
print(y.shape)
N, d = X.shape
```

<h2>Разбиваем данные на обучающую и тестовую выборки:</h2>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

<h2>Запускаем метод Random Forest:</h2>

Я выбрал Random Forest, потому что мой датасет содержит множество количественных и категориальных признаков, а также несбалансированные классы (63/37) с которыми этот метод автоматически справляется.

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

```python
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

print("Feature ranking:")

for f in range(d):
    print("%2d. feature '%5s' (%f)" % (f + 1, feature_names[indices[f]], importances[indices[f]]))  
```

```python
feature_names[[1, 0, 1]]
```

```python
d_first = 20
plt.title("Feature importances")
plt.bar(range(d_first), importances[indices[:d_first]], align = 'center', color = 'r')
plt.xticks(range(d_first), feature_names[indices[:d_first]], rotation = 90)
plt.xlim([-1, d_first])
```

Значения `err_train` и `err_test` равны 0.0, это означает, что модель предсказала все значения целевой переменной абсолютно точно как на тренировочной, так и на тестовой выборках. Признаки `reservation_status_Check-Out` и `reservation_status_Canceled` имеют наибольшую важность, потому что они напрямую указывают на статус бронирования, что почти определяет значение целевой переменной `is_canceled`. Как только модель замечает один из этих признаков, она легко и с высокой уверенностью делает предсказание о значении `is_canceled`, что приводит к нулевой ошибке на тренировочной и тестовой выборках.

Теперь попробуем запустить метод без этих признаков:

```python
new_X = data.drop(['reservation_status_Check-Out', 'reservation_status_Canceled','is_canceled'], axis=1)
new_X = new_X.to_numpy()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

```python
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

print("Feature ranking:")

for f in range(d - 2):
    print("%2d. feature '%5s' (%f)" % (f + 1, feature_names[indices[f]], importances[indices[f]]))  
```

```python
d_first = 20
plt.title("Feature importances")
plt.bar(range(d_first), importances[indices[:d_first]], align = 'center', color = 'r')
plt.xticks(range(d_first), feature_names[indices[:d_first]], rotation = 90)
plt.xlim([-1, d_first])
```

Ошибки 0.0 на тренировочной выборке и 0.091 на тестовой выборке указывают на улучшение поведения модели после удаления признаков `reservation_status_Check-Out` и `reservation_status_Canceled`. Теперь модель лучше отражает реальные зависимости в данных. Небольшое увеличение ошибки на тестовой выборке – это положительный признак, поскольку он говорит о том, что модель не просто "запомнила" данные, а нашла более общие паттерны.
